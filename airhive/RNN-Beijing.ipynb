```python
import tensorflow as tf
import pandas as pd
import numpy as np
import urllib
```

```python
tf.reset_default_graph()
```

```python
df_ = pd.read_csv("pechino.csv")
```

```python
df = pd.DataFrame(list(df_["pm2.5"]), columns = ["pm25"])
df["temp"] = df_.TEMP
df["pres"] = df_.PRES
df["prec"] = df_.Ir
df["vento"] = df_.Iws
df["mesi"] = df_.month
#df = df.join(pd.get_dummies(df_.month))
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
test = df.iloc[-301:]
df = df.iloc[:-100]
```

```python
n_steps = 700
n_inputs = len(df.columns)
n_neurons = 200
n_outputs = 1

len_layer = 2
```

```python
resolution = 1
lun = len(df) - n_steps

def time_series(t):
    array = []
    for x in t:
        array.append(df.iloc[x].values)
    return np.array(array).copy()

def next_batch(batch_size, n_steps):
    t0 = np.random.randint(lun, size = (batch_size, 1))
    Ts = t0 + np.arange(0, n_steps + 1) * resolution
    ys = time_series(Ts)
    return ys[:, :-1].reshape(-1, n_steps, n_inputs), ys[:, 1:, 0].reshape(-1, n_steps, n_outputs)
```

```python
X = tf.placeholder(tf.float32, [None, n_steps, n_inputs])
y = tf.placeholder(tf.float32, [None, n_steps, n_outputs])

cell = tf.contrib.rnn.OutputProjectionWrapper(
    tf.contrib.rnn.BasicRNNCell(num_units=n_neurons, activation=tf.nn.relu),
    output_size=n_outputs)

#cell = tf.nn.rnn_cell.DropoutWrapper(cell, 0.9)

outputs, states = tf.nn.dynamic_rnn(cell, X, dtype=tf.float32)
```

```python
learning_rate = 0.1

loss = tf.reduce_mean(tf.square(outputs - y))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
training_op = optimizer.minimize(loss)

init = tf.global_variables_initializer()
```

```python
saver = tf.train.Saver()
```

```python
n_iterations = 14000
batch_size = 50

with tf.Session() as sess:
    init.run()
    for iteration in range(n_iterations):
        X_batch, y_batch = next_batch(batch_size, n_steps)
        sess.run(training_op, feed_dict={X: X_batch, y: y_batch})
        if iteration % 100 == 0:
            mse = loss.eval(feed_dict={X: X_batch, y: y_batch})
            print(iteration, "\tMSE:", mse)
    
    saver.save(sess, "./modello_RNN_pechino")
```

```python
test_X = np.array(test.iloc[-(n_steps + 1):-1].values)
test_Y = np.array(test.pm25.iloc[-(n_steps):].values)

with tf.Session() as sess:
    saver.restore(sess, "./modello_RNN_pechino")

    X_new = test_X.reshape(1, -1, n_inputs)
    y_pred = sess.run(outputs, feed_dict={X: X_new})
```
